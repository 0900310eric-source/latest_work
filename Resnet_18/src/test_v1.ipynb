```python
import torch
import torchvision
# 引入 transform 處理圖片尺寸
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
from torchvision.models import resnet18, ResNet18_Weights
import torch.optim as optim
import torch.nn.functional as func

i = 3 # 很重要, 負責記錄 run i 的 log and pt
lr = 1e-3
torch.manual_seed(42) # 在程式最開頭加上這行

# 偵測有沒有 GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"目前使用的裝置: {device}")
# load models
model = resnet18(weights=ResNet18_Weights.DEFAULT)
# Fine tuning 前先換最後一層接口數
num_features = model.fc.in_features
model.fc = torch.nn.Linear(num_features, 10) # 作者最後一層就叫做 fc
model = model.to(device=device)

# 設定 optim
optimizer = optim.SGD(model.parameters(), lr=lr)
# 訓練次數
epochs = 20
# batch_size
batch_size = 128

# 自訂資料集
class MyDataSet(Dataset):
  def __init__(self, train=True):
    self.data = torchvision.datasets.CIFAR10(
            root='./data', 
            train=train, 
            download=True
        )
    # 定義轉換：將圖片轉成 Tensor，並縮放至 32x32
    self.transform = transforms.Compose([
          transforms.Resize((32, 32)),
          transforms.ToTensor(),
    ])

  def __len__(self):
    return len(self.data)

  def __getitem__(self, idx):
    # 取出 PIL 圖片與整數標籤 (0-9)
    img, label = self.data[idx]
    # 轉換圖片維度 [3, 256, 256]
    img_tensor = self.transform(img)
    prob = torch.zeros((10,))
    prob[label] = 1.0
    return img_tensor, prob

mydataset = MyDataSet(train=True)
# DataLoader
mydataloader = DataLoader(mydataset, batch_size=batch_size, shuffle=True)


print(f"Start training ...")
# 進入訓練
# 用 log 紀錄訓練過程
log_file = open(f"resnet18_cifar10_train_run{i}_log.txt", 'w')

for epoch in range(epochs):

  epoch_loss = 0.0
  for batch_idx, (batch_data, batch_p_probs) in enumerate(mydataloader):

    # ***** 把數據搬到 GPU，否則會跟 model 的位置對不上 *****
    batch_data = batch_data.to(device)
    batch_p_probs = batch_p_probs.to(device)

    # 在 Backward 之前記得清空舊的梯度 (這很重要！)
    optimizer.zero_grad()
    # forward pass
    prediction = model(batch_data)

    x_max, _ = torch.max(prediction, dim=1, keepdim=True)
    exp_prediction = torch.exp(prediction - x_max)
    exp_pred_sum = exp_prediction.sum(dim=1, keepdim=True)
    q_probs = exp_prediction / exp_pred_sum

    cross_entropy = -(batch_p_probs * torch.log(q_probs + 1e-7)).sum(dim=1)
    # 為了要讓 cross entropy 更穩定 (loss 穩定)
    loss = cross_entropy.mean()
    epoch_loss += loss.item()
    
    loss.backward()
    # 更新權重
    optimizer.step()
  
  avg_epoch_loss = epoch_loss / len(mydataloader)
  log_content = f"{avg_epoch_loss}\n"
  # 寫入 log.txt
  log_file.write(log_content)
  
  if (epoch + 1) % 5 == 0 or epoch == 0:
    print(avg_epoch_loss)

log_file.close() # 訓練完關閉檔案
print(f"End training ...")

# ***** 存 pt 檔 *****
# 設定存檔路徑
PATH = f"resnet18_cifar10_run{i}.pt"
torch.save(model.state_dict(), PATH)
```

```python
# 1. 先宣告一個空的 ResNet-18 (別忘了換頭！)
loaded_model = resnet18()
num_features = loaded_model.fc.in_features
loaded_model.fc = torch.nn.Linear(num_features, 10)

# 2. 把存好的 .pt 檔塞進去
loaded_model.load_state_dict(torch.load("resnet18_cifar10_run3.pt"))
loaded_model.to(device)
loaded_model.eval() # 進入評估模式
print("模型讀取完畢，隨時可以使用！")

def check_accuracy(loader, model):
    num_correct = 0
    num_samples = 0
    model.eval() # 評估模式
    
    with torch.no_grad():
        for x, y in loader:
            x, y = x.to(device), y.to(device)
            scores = model(x)
            # 因為 y 是 one-hot，先轉回 index
            _, labels = torch.max(y, 1) 
            _, predictions = scores.max(1)
            
            num_correct += (predictions == labels).sum()
            num_samples += predictions.size(0)
        
    acc = float(num_correct) / num_samples * 100
    print(f"在 {num_samples} 張測試圖中，準確率為: {acc:.2f}%")
    return acc

# 建立測試用的 Loader
test_dataset = MyDataSet(train=False) # 記得你的 Dataset 要支援 train=False
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# 開始體檢
check_accuracy(test_loader, loaded_model)
```

```python
# 畫 train loss 比較圖
import matplotlib.pyplot as plt
import numpy as np

# --- 字體設定 ---
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman"] + plt.rcParams["font.serif"]
# 如果電腦是 Linux 或 Colab，有時需要這行來處理數學符號
plt.rcParams['mathtext.fontset'] = 'stix' 
# -----------------------

log_paths = ['resnet18_cifar10_train_run1_log.txt', 'resnet18_cifar10_train_run2_log.txt', 'resnet18_cifar10_train_run3_log.txt']
loss_per_run = []

for log in log_paths:
  with open(log, 'r', encoding='utf-8') as f:
    loss_data = f.read()
    temp_loss = loss_data.strip().split('\n')
    temp_loss = [float(loss) for loss in temp_loss]
    loss_per_run.append(temp_loss)
    # print(temp_loss)

loss_per_run_array = np.array(loss_per_run, np.float32)
# 設定圖的大小
plt.figure(figsize=(12, 6.75))
for i in range(loss_per_run_array.shape[0]):
  plt.plot(loss_per_run_array[i, :], label=f"Run {i+1}")

# 設定圖表細節
# plt.title("Training Loss Comparison")
plt.xlabel("Epochs")
plt.ylabel("Training loss")
plt.legend() # 顯示右上角的標籤
# plt.grid(True) # 顯示格線
plt.show()
```
